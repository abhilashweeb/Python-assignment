```python
#7.  Write a program that performs basic arithmetic operations on two user-input numbers.
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Function to add two numbers
def addition(n1, n2):
    print(f"{n1} + {n2} = {n1 + n2}")

# Function to multiply two numbers
def multiplication(n1, n2):
    print(f"{n1} * {n2} = {n1 * n2}")

# Function to subtract two numbers
def subtraction(n1, n2):
    print(f"{n1} - {n2} = {n1 - n2}")

# Function to divide two numbers
def division(n1, n2):
    if n2 == 0:
        print("Division by zero is not possible")
    else:
        print(f"{n1} / {n2} = {n1 / n2}")

# Calling the functions
addition(num1, num2)
multiplication(num1, num2)
subtraction(num1, num2)
division(num1, num2)
```

```python
#8. Write a program to demonstrate the use of logical operators.
# Logical 'and'
age = int(input("Enter your age: "))
has_id = input("Do you have a voter ID card? (yes or no): ").strip().lower()

def verify(age, has_id):
  if age >= 18 and has_id == "yes":
    print("You are eligible to vote")
  else:
    print("You are not eligible to vote")

verify(age, has_id)
# Here logical "and" checks if both the conditions are true it will execute the if block of code.
```

```python
# Logical 'or'
age = int(input("Enter your age: "))
has_id = input("Do you have a voter ID card? (yes or no): ").strip().lower()
def verify_with_or(age, has_id):
  if age >= 18 or has_id == "yes":
    print("You are eligible to vote")
  else:
    print("You are not eligible to vote")

verify_with_or(age, has_id)
# The logical  "or" operator checks if at least one condition is true if so, it executes thr if block code.
```

```python
#10. Write code to demonstrate type casting with list elements.
str_list = ["10", "15.5", "True", "0", "False"]
int_list = [int(i) if i.isdigit() else 0 for i in str_list] #Default to 0 if not an integer
float_list = [float(i) if '.' in i or i.replace('.', '', 1).isdigit() else 0.0 for i in str_list]
bool_list = [bool(i.strip()) for i in str_list]
#Output results
print("Original List:", str_list)
print("Integer List:", int_list)
print("Float List:", float_list)
print("Boolean List:", bool_list)
```

```python
#14. Write a program to reverse a string using a while loop.
string = input("Enter a string: ")

reversed_string = ""
index = len(string) - 1

while index >= 0:
    reversed_string += string[index]
    index -= 1

print("Reversed string:", reversed_string)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
number = int(input("Enter a number to find factorial: "))
result = 10
i = 5
while i <= number:
  result *= i
  i += 5
  print(f"Factorial of {number} = {result}")
```
